```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder as le
```

```python
df = pd.read_csv('../datasets/train.csv')
```

```python
df.head()
```

```python
## codifica cada nome diferente como um número
name_encoder = le()
feature = 'nome'
name_encoder.fit(df[feature])
df[feature] = name_encoder.transform(df[feature])
```

```python
def column2dummy(dataframe, column_name):
    '''
    Função que transforma uma coluna de tipo texto ou fator em variáveis dummy
    '''
    values = le()
    values = values.fit(dataframe[column_name])
    values = values.classes_
    for i in range(len(values) - 1):
        value = values[i]
        col_str = column_name + '_' + value
        dataframe[col_str] = np.where(df[column_name].str.contains(value), 1, 0)
        
    return dataframe.drop(column_name, axis = 1)
```

```python
to_dummy = ['uf', 'partido', 'cargo', 'estado_civil', 'ocupacao', 'sexo']
for variable in to_dummy:
    df = column2dummy(df, variable)
    
## Como grau é uma variavel que pode ser ordenada, será feito um label encoding
## de modo que 0 seja o menor grau de escolaridade, e cresça a cada nivel

# grau_encoding = le()
# grau_encoding = grau_encoding.fit(df['grau'])
# print grau_encoding.classes_

niv_escolaridade = {'LÊ E ESCREVE': 0, 'ENSINO FUNDAMENTAL INCOMPLETO': 1, 'ENSINO FUNDAMENTAL COMPLETO': 2,
'ENSINO MÉDIO INCOMPLETO': 3, 'ENSINO MÉDIO COMPLETO': 4, 'SUPERIOR INCOMPLETO': 5, 'SUPERIOR COMPLETO': 6}

ordenar_grau = []

for i, row in df.iterrows():
    ordenar_grau.append(niv_escolaridade[row['grau']])
    
df['grau'] = ordenar_grau
```

```python
df.to_csv('../datasets/train_factors.csv')
```
